## Reading a html file: 

```python
from bs4 import BeautifulSoup

with open("index.html", "r") as f: 
    doc = BeautifulSoup(f,"html.parser")
print(doc.prettify())
```

## Find by tag name: 
You can access a tag in beautiful soup using doc.tagname, this gives you access to the first occurence of the tag. 

```python
print(doc.title)
print(doc.h1)
print(doc.p)
# or we can use .find()
print("\n", "Or we can use .find() : \n")
print(doc.find("title"))
print(doc.find("h1"))
print(doc.find("p"))
```

```python
tag = doc.find(class_ = "apprenez")
print(tag)
```

### Selecting tags that its text has a certain character or word:

Earlier we tried to select the price of the iphone but to do that we put in too much effort, there is an easier way to do that:

```python
import re
tag = html_code.find_all("span",text = re.compile("DT.*")) #. means anything after DT and * means any number of things
print(tag[1].text)
```

### The tree structure of Html: 
Each HTML document can actually be referred to as a document tree. We describe the elements in the tree like we would describe a family tree. There are ancestors, descendants, parents, children and siblings.

Suppose this html code :


    

```python
html = "<a> This is the very first parent <b> this is its first child, and it's c's previous sibling </b> <c> this is b's next sibling  <d> d is c's child and a's descent </d></c><a/>"
soup = BeautifulSoup(html, "html.parser")
print(soup.b)
for element in soup.b.next_siblings:
    print(element)
```

```python
for element in soup.c.children: 
    print(element)
```

```python
for element in soup.a.descendants: 
    print(element)
```

```python
url = "https://www.tunisianet.com.tn/301-pc-portable-tunisie?page=2&order=product.price.asc"
result = requests.get(url)
soup = BeautifulSoup(result.text,"html.parser")
print(soup.prettify())
```

```python
tags = soup.find_all(["h2"], class_ = "product-title")
my_list = []
for tag in tags: 
    title = tag.text
    L = title.split("/")
    d = {"Title" : L[0], "Processor" : L[1], "Ram" : L[2]}
    my_list.append(d)
my_list
from pandas import DataFrame
df = DataFrame(my_list)
df
    
```
